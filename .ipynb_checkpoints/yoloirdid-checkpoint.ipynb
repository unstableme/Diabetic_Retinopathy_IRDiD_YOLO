```python
import os
import cv2
import numpy as np
from PIL import Image
from collections import defaultdict

# === PATHS FOR TRAIN SET ===
image_dir = r"C:\USERS\USER\DOCUMENTS\FIRST YOLO_DIABETIC RETINOPATHY\data\images\train"
mask_root_dir = r"C:\Users\user\Documents\First YOLO_Diabetic Retinopathy\A. Segmentation\2. All Segmentation Groundtruths\a. Training Set"
output_label_dir = r"C:\USERS\USER\DOCUMENTS\FIRST YOLO_DIABETIC RETINOPATHY\data\labels\train"

os.makedirs(output_label_dir, exist_ok=True)


class_folders = {
    "1. Microaneurysms": 0,
    "2. Haemorrhages": 1,
    "3. Hard Exudates": 2,
    "4. Soft Exudates": 3
}

merged_labels = defaultdict(list)

# === PROCESS EACH MASK FOLDER ===
for lesion_folder, class_id in class_folders.items():
    mask_folder = os.path.join(mask_root_dir, lesion_folder)
    print(f"Processing: {lesion_folder}...")

    for filename in os.listdir(mask_folder):
        if not filename.lower().endswith(".tif"):
            continue

        image_id = "_".join(filename.split("_")[:2])  # IDRiD_01
        image_name = image_id + ".jpg"
        image_path = os.path.join(image_dir, image_name)

        if not os.path.exists(image_path):
            print(f"Skipped: {image_name} not found in images/train/")
            continue

        img = Image.open(image_path)
        img_w, img_h = img.size

        mask_path = os.path.join(mask_folder, filename)
        mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)

        if mask is None:
            print(f"Could not load: {filename}")
            continue

        # Ensure 3 channels
        if len(mask.shape) == 2:
            mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
        elif mask.shape[2] == 4:
            mask = cv2.cvtColor(mask, cv2.COLOR_BGRA2BGR)

        hsv = cv2.cvtColor(mask, cv2.COLOR_BGR2HSV)

        # Red detection (loosened range for faint dots)
        lower_red = np.array([0, 0, 50])
        upper_red = np.array([180, 255, 255])
        binary = cv2.inRange(hsv, lower_red, upper_red)

        contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            if w < 3 or h < 3:
                continue

            x_center = (x + w / 2) / img_w
            y_center = (y + h / 2) / img_h
            w_norm = w / img_w
            h_norm = h / img_h

            merged_labels[image_id].append(f"{class_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}")


for image_id, lines in merged_labels.items():
    label_path = os.path.join(output_label_dir, image_id + ".txt")
    with open(label_path, "w") as f:
        f.write("\n".join(lines))

print("YOLO labels generated and merged for training set!")
```

```python

# === PATH SETUP ===
image_dir = r"C:\USERS\USER\DOCUMENTS\FIRST YOLO_DIABETIC RETINOPATHY\data\images\val"
mask_root_dir = r"C:\USERS\USER\DOCUMENTS\FIRST YOLO_DIABETIC RETINOPATHY\A. Segmentation\2. All Segmentation Groundtruths\b. Testing Set"
output_label_dir = r"C:\USERS\USER\DOCUMENTS\FIRST YOLO_DIABETIC RETINOPATHY\data\labels\val"

os.makedirs(output_label_dir, exist_ok=True)

# === CLASS MAPPING ===
class_folders = {
    "1. Microaneurysms": 0,
    "2. Haemorrhages": 1,
    "3. Hard Exudates": 2,
    "4. Soft Exudates": 3
}

from collections import defaultdict

merged_labels = defaultdict(list)

# === PROCESS EACH MASK FOLDER ===
for lesion_folder, class_id in class_folders.items():
    mask_folder = os.path.join(mask_root_dir, lesion_folder)
    print(f"Processing: {lesion_folder}...")

    for filename in os.listdir(mask_folder):
        if not filename.lower().endswith(".tif"):
            continue

        image_id = "_".join(filename.split("_")[:2])  # IDRiD_55
        image_name = image_id + ".jpg"
        image_path = os.path.join(image_dir, image_name)

        if not os.path.exists(image_path):
            print(f"Skipped: {image_name} not found in images/val/")
            continue

        # Load image for size
        img = Image.open(image_path)
        img_w, img_h = img.size

        # Read and clean mask
        mask_path = os.path.join(mask_folder, filename)
        mask = cv2.imread(mask_path, cv2.IMREAD_UNCHANGED)

        if mask is None:
            print(f"Could not load: {filename}")
            continue

        if len(mask.shape) == 2:
            mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
        elif mask.shape[2] == 4:
            mask = cv2.cvtColor(mask, cv2.COLOR_BGRA2BGR)

        hsv = cv2.cvtColor(mask, cv2.COLOR_BGR2HSV)

        # For very faint/gray-red ‚Üí broaden filter range
        lower_red = np.array([0, 0, 50])
        upper_red = np.array([180, 255, 255])
        binary = cv2.inRange(hsv, lower_red, upper_red)

        contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        for cnt in contours:
            x, y, w, h = cv2.boundingRect(cnt)
            if w < 3 or h < 3:
                continue

            x_center = (x + w / 2) / img_w
            y_center = (y + h / 2) / img_h
            w_norm = w / img_w
            h_norm = h / img_h

            merged_labels[image_id].append(f"{class_id} {x_center:.6f} {y_center:.6f} {w_norm:.6f} {h_norm:.6f}")

# üíæ Save merged labels
for image_id, lines in merged_labels.items():
    label_path = os.path.join(output_label_dir, image_id + ".txt")
    with open(label_path, "w") as f:
        f.write("\n".join(lines))

print("All labels merged and saved in YOLO format!")

```

```python
yaml_content = """\
train: C:/USERS/USER/DOCUMENTS/FIRST YOLO_DIABETIC RETINOPATHY/data/images/train
val: C:/USERS/USER/DOCUMENTS/FIRST YOLO_DIABETIC RETINOPATHY/data/images/val
nc: 4
names: ["Microaneurysms", "Haemorrhages", "Hard Exudates", "Soft Exudates"]
"""

with open("data.yaml", "w") as f:
    f.write(yaml_content)

print("data.yaml created!")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data ='data.yaml',
    batch=8,
    epochs=2,
    imgsz=640,
    name="yolov8_diabetic_retinopathy"
)
```
